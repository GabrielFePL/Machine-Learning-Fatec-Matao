## 6. Impacto da Renda no Score de Crédito

### Verifique se há uma relação direta entre renda e pontuação de crédito.

```python
def score_calculate(income, age, loan, default):
    Indebtedness = loan / income
    score = (1000 - (11000 * Indebtedness**1.5) + (age * 1.25) - (200 * default))
    return max(min(score, 1000), 0)
```

```python
credit_dataframe['score'] = credit_dataframe.apply(lambda row: score_calculate(row['income'], row['age'], row['loan'], row['default']), axis=1)
```

## 10. Conclusões e Recomendações para Modelagem

### Com base nos insights obtidos, quais variáveis parecem mais relevantes para um modelo de aprendizado de máquina?

As variáveis que se sobressaem como potencialmente relevantes para um modelo de aprendizado de máquina são a renda, devido à correlação observada entre menores rendimentos e pontuações de crédito reduzidas; o empréstimo, já que a relação entre o valor do empréstimo e a renda demonstrou ser um fator crucial — valores mais altos em comparação à renda levaram a scores inferiores; e a idade, que, como indicador do tempo de histórico de crédito, evidenciou que clientes mais jovens apresentam maior índice de inadimplência.

### Há necessidade de engenharia de features ou tratamento de dados para melhorar a qualidade do modelo?

Para aprimorar a qualidade do modelo, é essencial aplicar técnicas de engenharia de features e um tratamento adequado dos dados. A criação de variáveis derivadas pode contribuir para um desempenho superior, como a taxa de endividamento, que representa a proporção entre o valor do empréstimo e a renda do indivíduo. Além disso, a categorização da idade em faixas específicas (por exemplo, 20-30, 30-40, etc.) pode ajudar a capturar padrões relevantes. Um score ajustado, considerando a interação entre idade, renda e empréstimo, também pode melhorar a capacidade preditiva do modelo, conforme identificado em análises anteriores.

Outro ponto importante é o tratamento de valores ausentes e anômalos na variável idade, garantindo um aprendizado mais eficiente. Na engenharia de features, pode-se calcular a relação entre empréstimo e renda para avaliar o nível de endividamento e categorizar a idade para entender melhor sua influência no histórico de crédito e na inadimplência. Além disso, a exploração de interações entre variáveis por meio de operações matemáticas, como multiplicação e divisão, pode revelar relações não lineares. Técnicas de normalização e escalonamento também são indispensáveis, especialmente para modelos sensíveis a variações de escala, como a Regressão Logística.
